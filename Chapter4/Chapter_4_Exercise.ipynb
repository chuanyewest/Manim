`Exercise 1` **Solution**

```python
#Solution to the Exercise (Hidden for Students)
# Only reveal this if needed for teaching purposes
%%manim -qm ShapeDanceSolution

config.media_dir = "/content/media"

class ShapeDanceSolution(Slide):
    def construct(self):
        # Slide 1: Create a triangle
        triangle = Triangle(color=RED)
        self.play(Create(triangle))
        self.next_slide()

        # Slide 2: Transform to a pentagon
        pentagon = RegularPolygon(n=5, color=GREEN)
        self.play(Transform(triangle, pentagon))
        self.next_slide()

        # Slide 3: Scale up and down in a loop
        self.play(Rotate(pentagon, angle=2*PI, run_time=2, rate_func=linear))
        self.next_slide(loop=True)  # Loop until manually advanced

        # Slide 4: Add text and end
        text = Text("Shape Dance Complete!", color=YELLOW).shift(DOWN)
        self.play(Write(text))
        self.next_slide()
```

`Exercise 2` **Solution**

Here, we use `animate.shift(direction * distance))` as an alternative approach for `shift(direction * distance)` to control the time

```python
%%manim -qm ArrowJourney

config.media_dir = "/content/media"

class ArrowJourney(Slide):
    def construct(self):
        # Slide 1: Green arrow pointing right
        arrow = Arrow(start=LEFT, end=RIGHT, color=GREEN)
        self.play(Create(arrow))
        self.next_slide()

        # Slide 2: Flip to left and turn red
        flipped_arrow = Arrow(start=RIGHT, end=LEFT, color=RED)
        self.play(Transform(arrow, flipped_arrow))
        self.next_slide()

        # Slide 3: Move up and down
        self.play(arrow.animate.shift(UP), run_time=0.5)
        self.play(arrow.animate.shift(DOWN), run_time=0.5)
        self.play(arrow.animate.shift(UP), run_time=0.5)
        self.play(arrow.animate.shift(DOWN), run_time=0.5)
        self.wait(1)
```

```python
%%manim -qm IntegrationVisual

class IntegrationVisual(Slide):
    def construct(self):
        # Slide 1: Plot the function f(x) = x^2 from x = 0 to x = 2
        axes = Axes(
            x_range=[0, 2.5, 0.5],
            y_range=[0, 5, 1],
            axis_config={"color": WHITE},
        ).add_coordinates()
        func = axes.plot(lambda x: x**2, x_range=[0, 2], color=BLUE)
        func_label = MathTex("f(x) = x^2")
        func_label.shift(UP*3)

        # func_label = MathTex("f(x) = x^2").next_to(axes, UP, buff=0.5)

        self.add(axes, func_label)
        self.play(Write(func),run_time=1)
        self.wait(1)
        self.next_slide()

        # Slide 2: Approximate the area with 4 rectangles (Riemann sum)
        rectangles = axes.get_riemann_rectangles(
            func,
            x_range=[0, 2],
            dx=0.5,  # 4 rectangles (2 / 0.5 = 4)
            color=ORANGE,
            fill_opacity=0.5,
        )
        self.play(Create(rectangles), run_time=2)
        self.wait(1)
        self.next_slide()

        # Slide 3: Transform rectangles into exact area and animate the integral equation
        exact_area = axes.get_area(func, x_range=[0, 2], color=GREEN, opacity=0.5)
        self.play(Transform(rectangles, exact_area), run_time=2)

        # Create and animate the integral equation
        integral = MathTex(r"\int_{0}^{2} x^2 \, dx")

        self.play(Write(integral), run_time=2)

        # Final label
        complete_label = Text("Integration Complete!", color=YELLOW).to_edge(DOWN)
        self.play(Write(complete_label), run_time=1)
        self.wait(2)
```
